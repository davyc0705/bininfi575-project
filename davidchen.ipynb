#### I did some coding below to read the documents we need in the project, and I have got rid of the columns we may not use.

```python
column_name = ['CHROM', 'POS', 'ID', 'REF', 'ALT', 'QUAL', 'FILTER', 'INFO','Format' ,'Group-c']

def readfile(filename):
    read_file = pd.read_csv(filename, comment = '#', sep = '	', header = None)
    read_file.columns = column_name
    read_file = read_file.loc[:,'CHROM':'ALT'].drop('ID',1)
    return read_file.head()
```

```python
readfile('TNBC.vcf')
```

```python
readfile('nTNBC.vcf')
```

```python
readfile('HER2.vcf')
```
